```python
import matplotlib.pyplot as plt
import seaborn as sns

# Countplot
plt.figure(figsize=(6, 4))
sns.countplot(x='ACTION', data=df)
plt.title("Distribution of Target Variable (ACTION)")
plt.xlabel("ACTION")
plt.ylabel("Count")
plt.tight_layout()

# Save the plot
plt.savefig("outputs/eda_target_distribution.png")


plt.show()
```

```python
unique_counts = df.nunique().sort_values()
print(unique_counts)
```

```python
# Select categorical features (excluding ID and ACTION)
cat_features = [col for col in df.columns if col not in ['ID', 'ACTION']]

# Plot top 10 values for each
for col in cat_features:
    plt.figure(figsize=(8, 3))
    sns.countplot(x=col, data=df, order=df[col].value_counts().index[:10])
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Save each plot
    plt.savefig(f"outputs/eda_dist_{col}.png")
    plt.show()
```
